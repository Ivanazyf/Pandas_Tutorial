## Question: Which continent grows fastest?

```python
df.groupby('Continent').mean().sort_values('2022 Population', ascending = False)
```

```python
# Pull out rows containing 'Oceania'

df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean().sort_values('2022 Population', ascending = False)
df2
```

```python
df2 = df2.drop(columns = ['Rank','Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage'])
```

```python
# Reverse column orders

df2 = df2[df2.columns[::-1]]
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
